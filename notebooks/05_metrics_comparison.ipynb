```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# Synthetic dataset
np.random.seed(42)
X = 2 * np.random.rand(80, 1)
y = 3 + 2 * X + np.random.randn(80, 1)

# Add an outlier
X[0] = 1.8
y[0] = 15

# Train model
model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)

# Metrics
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y, y_pred)
r2 = r2_score(y, y_pred)

print("MSE :", mse)
print("RMSE:", rmse)
print("MAE :", mae)
print("R²  :", r2)
```

```python
# Plot
plt.figure()
plt.scatter(X, y, label="Data")
plt.plot(X, y_pred, label="Regression Line")
plt.legend()
plt.title("Regression with Outlier")
plt.show()
```

```python
# Remove outlier
X_clean = X[1:]
y_clean = y[1:]

model.fit(X_clean, y_clean)
y_pred_clean = model.predict(X_clean)

mse_clean = mean_squared_error(y_clean, y_pred_clean)
mae_clean = mean_absolute_error(y_clean, y_pred_clean)

print("\nWithout Outlier:")
print("MSE:", mse_clean)
print("MAE:", mae_clean)
```

**MSE penalizes large errors more heavily due to squaring, making it sensitive to outliers. MAE provides a more robust measure under noisy or heavy-tailed error distributions.**

**RMSE is often preferred for interpretability since it shares the same units as the target variable. MAE is more robust to outliers, while MSE emphasizes large deviations. R^2 quantifies variance explained by the model**

# Regression Metrics Comparison

## Objective
This notebook compares common regression evaluation metrics:

- MSE
- RMSE
- MAE
- R² Score

And examines their sensitivity to outliers.

---

## Dataset
Synthetic linear dataset with Gaussian noise.

An artificial outlier was introduced to assess metric behavior.

---

## Metrics Explained

### MSE (Mean Squared Error)
- Penalizes large errors strongly
- Sensitive to outliers

---

### RMSE (Root Mean Squared Error)
- Square root of MSE
- Same units as target

---

### MAE (Mean Absolute Error)
- Linear penalty
- More robust to outliers

---

### R² Score
- Fraction of variance explained
- Measures goodness of fit

---

## Results & Interpretation
- Outlier significantly increases MSE
- MAE less dramatically affected
- Regression line shifts toward outlier

---

## Key Insight
Choice of metric influences model evaluation, especially in noisy or heavy-tailed datasets.

---

## Conclusion
Different metrics emphasize different aspects of prediction error and should be selected based on problem context.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
#simulate data
np.random.seed(0)
X = np.random.randn(100, 2)
y = (X[:, 0] + X[:, 1] > 0).astype(int)

#train model
model = LogisticRegression()
model.fit(X, y)

#predictions
y_pred = model.predict(X)
```

```python
#confusion matrix
cm = confusion_matrix(y, y_pred)
cm
```

```python
#metrics
acc = accuracy_score(y, y_pred)
prec = precision_score(y, y_pred)
rec = recall_score(y, y_pred)
f1 = f1_score(y, y_pred)

print("Accuracy :", acc)
print("Precision:", prec)
print("Recall   :", rec)
print("F1 Score :", f1)
```

```python
plt.figure()
plt.imshow(cm)
plt.title("Confusion Matrix")
plt.colorbar()
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

# Confusion Matrix & Classification Metrics

## Objective
The goal of this notebook is to evaluate a binary classification model using a confusion matrix and common classification performance metrics.

---

## Dataset
We use a synthetic binary classification dataset with two input features and linearly separable classes.

---

## Model
A Logistic Regression classifier is trained to predict class membership.

---

## Evaluation Tools

### Confusion Matrix
Summarizes prediction outcomes:

\[
\begin{bmatrix}
TN & FP \\
FN & TP
\end{bmatrix}
\]

Where:

- TN → True Negatives  
- FP → False Positives  
- FN → False Negatives  
- TP → True Positives  

---

### Metrics

**Accuracy**  
Fraction of correct predictions.

**Precision**  
Among predicted positives, how many are correct.

**Recall**  
Among true positives, how many were detected.

**F1 Score**  
Harmonic mean of Precision and Recall.

---

## Key Insight
Different metrics emphasize different aspects of model performance. Accuracy alone may be misleading, especially in imbalanced datasets.

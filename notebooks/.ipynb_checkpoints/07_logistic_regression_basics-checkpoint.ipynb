```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
# Synthetic dataset
np.random.seed(0)
X = np.random.randn(100, 2)

# Define labels (simple boundary)
y = (X[:, 0] + X[:, 1] > 0).astype(int)

# Train model
model = LogisticRegression()
model.fit(X, y)

# Create grid for decision boundary
xx, yy = np.meshgrid(
    np.linspace(X[:,0].min()-1, X[:,0].max()+1, 100),
    np.linspace(X[:,1].min()-1, X[:,1].max()+1, 100)
)

grid = np.c_[xx.ravel(), yy.ravel()]
probs = model.predict_proba(grid)[:, 1]
probs = probs.reshape(xx.shape)

# Plot
plt.figure()
plt.contourf(xx, yy, probs, alpha=0.3)
plt.scatter(X[:,0], X[:,1], c=y)
plt.title("Logistic Regression Decision Boundary")
plt.show()
```

```python
sample_probs = model.predict_proba(X[:5])
print("Sample Probabilities:\n", sample_probs)
```

**A grid of feature values is generated to evaluate model probabilities across input space. The contour plot visualizes predicted class probabilities, while the decision boundary corresponds to regions where the predicted probability equals 0.5.**

# Logistic Regression & Decision Boundaries

## Objective
The purpose of this notebook is to introduce **binary classification** using Logistic Regression and to build intuition for probabilistic predictions and decision boundaries.

We demonstrate:

- Synthetic data generation
- Binary label construction
- Logistic Regression training
- Class probability prediction
- Decision boundary visualization

---

## Dataset
A synthetic two-dimensional dataset was generated:

- 100 samples
- 2 input features per sample
- Normally distributed feature values

Class labels were defined using a linear rule:

\[
y =
\begin{cases}
1 & \text{if } x_1 + x_2 > 0 \\
0 & \text{otherwise}
\end{cases}
\]

This produces two linearly separable classes.

---

## Model
We use **Logistic Regression**, a linear classifier that models the probability of class membership:

\[
P(y=1 \mid X) = \sigma(z)
\]

where

\[
z = w_1 x_1 + w_2 x_2 + b
\]

and

\[
\sigma(z) = \frac{1}{1 + e^{-z}}
\]

Key characteristics:

- Outputs probabilities rather than continuous values
- Uses sigmoid activation
- Learns a linear decision boundary

---

## Methodology
1. Generate synthetic feature data  
2. Construct binary labels  
3. Fit Logistic Regression model  
4. Predict class probabilities  
5. Evaluate model across feature space  
6. Visualize probability contours and decision boundary  

---

## Results & Interpretation
The contour plot illustrates:

- Smooth probability transitions between classes
- Linear separation learned by the classifier
- Decision boundary occurring near \( P = 0.5 \)

Observed behavior:

- Regions of high probability correspond to Class 1
- Boundary reflects classifier uncertainty

---

## Key Insight
Logistic Regression:

- Converts linear combinations of features into probabilities
- Provides interpretable decision boundaries
- Forms a baseline classifier for binary prediction tasks

---

## Conclusion
This notebook establishes foundational classification concepts including probabilistic prediction and decision boundary interpretation, which are essential for evaluating and diagnosing classification models.

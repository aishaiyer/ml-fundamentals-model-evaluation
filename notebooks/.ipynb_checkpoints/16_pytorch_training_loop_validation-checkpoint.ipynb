```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
#create dataset
torch.manual_seed(0)

X = torch.randn(300, 2)
y = (X[:, 0] + X[:, 1] > 0).float().view(-1, 1) #binary

#test/train split note that sklern needs numpy input
X_train, X_val, y_train, y_val = train_test_split(
    X.numpy(), y.numpy(), test_size=0.25, random_state=42 
)

#turning them back to torch
X_train = torch.tensor(X_train, dtype=torch.float32)
X_val   = torch.tensor(X_val, dtype=torch.float32)
y_train = torch.tensor(y_train, dtype=torch.float32)
y_val   = torch.tensor(y_val, dtype=torch.float32)
```

```python
#create model object again
class SimpleNN(nn.Module):
    def __init__(self):
        super(SimpleNN, self).__init__()
        self.layer1 = nn.Linear(2, 16)
        self.relu = nn.ReLU()
        self.layer2 = nn.Linear(16, 1)

    def forward(self, x):
        x = self.layer1(x)
        x = self.relu(x)
        x = self.layer2(x)
        return x

model = SimpleNN()
```

```python
#loss and optimizer, binary/sigmoid and learning rate set
criterion = nn.BCEWithLogitsLoss()
optimizer = optim.Adam(model.parameters(), lr=0.01) #using Adaptive moment estimation for faster convergence and adapting learning rate per parameter stably
```

```python
#training loop
train_losses = []
val_losses = []

epochs = 50

for epoch in range(epochs):

    # ---- TRAINING ----
    model.train() #enables training behavior
    outputs = model(X_train)
    train_loss = criterion(outputs, y_train)

    optimizer.zero_grad() #initalizes gradient by clearing old
    train_loss.backward() #backward propagation by computing gradients
    optimizer.step() #update weights

    # ---- VALIDATION ----
    model.eval() #switches to inference mode
    with torch.no_grad(): #stops gradient tracking to save memory & compute
        val_outputs = model(X_val)
        val_loss = criterion(val_outputs, y_val)

    train_losses.append(train_loss.item())
    val_losses.append(val_loss.item())

    print(f"Epoch {epoch+1:02d} | Train Loss = {train_loss.item():.4f} | Val Loss = {val_loss.item():.4f}")
```

```python
#plot loss curves, if both going down with epoch then good learning, if training curve going down but validation going up then overfitting, 
#and if both flat then underfitting
plt.figure()
plt.plot(train_losses, label="Train Loss")
plt.plot(val_losses, label="Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training vs Validation Loss")
plt.legend()
plt.show()
```

# Training Loop & Validation Monitoring

## Objective
This notebook implements a full PyTorch training loop while tracking training and validation loss to diagnose learning behavior.

---

## Key Concepts

**Training Mode (`model.train`)**  
Enables gradient-based parameter updates.

**Evaluation Mode (`model.eval`)**  
Disables training-specific behaviors.

**Validation Loss**  
Measures generalization performance.

---

## Key Insight
Comparing training and validation loss provides early detection of overfitting and training instability.

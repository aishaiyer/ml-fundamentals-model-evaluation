```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_recall_curve, average_precision_score
```

```python
#create imbalanced dataset
np.random.seed(42)

n_samples = 500
X = np.random.randn(n_samples, 2)

# Only 5% positives
y = (np.random.rand(n_samples) < 0.05).astype(int)
```

```python
#train model
model = LogisticRegression()
model.fit(X,y)
y_probs = model.predict_proba(X)[:, 1]
```

```python
#compute PR curve
precision, recall, thresholds = precision_recall_curve(y, y_probs)
ap = average_precision_score(y, y_probs)

print("Average Precision (AP):", ap)

#plot PR curve
plt.figure()
plt.plot(recall, precision, label=f"AP = {ap:.3f}")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision–Recall Curve")
plt.legend()
plt.show()
```

```python
#threshold behavior
for thresh in [0.1, 0.3, 0.5, 0.7]:
    y_pred_thresh = (y_probs > thresh).astype(int)
    print(f"\nThreshold {thresh}")
    print("Predicted positives:", y_pred_thresh.sum())
```

```python
#class weighting
model = LogisticRegression(class_weight='balanced')
model.fit(X, y)
#inspecting probabilities
print("Max probability:", y_probs.max())
print("Mean probability:", y_probs.mean())
```

# Precision–Recall Curve & Imbalanced Classification

## Objective
This notebook evaluates classifier performance on an imbalanced dataset using the Precision–Recall (PR) curve.

---

## Key Concepts

**Precision**  
Fraction of predicted positives that are correct.

**Recall**  
Fraction of actual positives correctly identified.

---

## Why Precision–Recall?
For imbalanced datasets, accuracy and ROC curves may present overly optimistic performance estimates. PR curves provide more informative evaluation when the positive class is rare.

---

## Average Precision (AP)
Summarizes the PR curve into a single performance metric. Higher values indicate better classifier discrimination for the positive class.
